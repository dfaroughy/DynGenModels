```python
import torch
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from DynGenModels.models.mlp import MLP
from DynGenModels.configs.toys_configs import Toys_FlowMatch_MLP_Configs as Configs

#...define hyperparameters:

configs = Configs(data_name = '8gauss_to_2moons',
                features = ['x', 'y'],
                num_points = 20000,
                sigma = 0.0001,
                gauss_8_scale = 5,
                gauss_8_var = 0.1,
                moon_2_noise = 0.1,
                data_split_fracs = [0.7, 0.2, 0.1],
                epochs = 1000,
                early_stopping=1000,
                batch_size = 256,
                warmup_epochs = 50,
                print_epochs = 100,
                lr = 1e-4,
                dim_hidden = 64, 
                device = 'cpu',
                solver='dopri5',
                num_sampling_steps=100,
                seed = 1234
                )

configs.set_workdir(path='../results', save_config=True)
```

```python
from DynGenModels.datamodules.toys.datasets import Gauss2MoonsDataset as DataSet
from DynGenModels.datamodules.toys.dataloader import ToysDataLoader as Loader

dataset = DataSet(configs)
dataloader = Loader(dataset, configs)
```

```python
from DynGenModels.dynamics.flowmatch import SimplifiedCondFlowMatching
from DynGenModels.trainer.trainer import DynGenModelTrainer

mlp = MLP(configs)
cfm = DynGenModelTrainer(dynamics=SimplifiedCondFlowMatching(mlp, configs), 
                         dataloader=dataloader,
                         configs=configs)

cfm.train()
```

```python
from DynGenModels.pipelines.SamplingPipeline import FlowMatchPipeline as Pipeline
pipeline = Pipeline(trained_model=cfm, configs=configs)
```

```python
fig, ax = plt.subplots(1,1, figsize=(4,4))
for i in range(configs.num_sampling_steps): 
    plt.scatter(pipeline.trajectories[i][:,0], pipeline.trajectories[i][:,1], s=0.05, color='gray', alpha=0.2)
plt.scatter(pipeline.trajectories[0][:,0], pipeline.trajectories[0][:,1], s=0.5, color='red')
plt.scatter(pipeline.trajectories[-1][:,0], pipeline.trajectories[-1][:,1], s=0.5, color='blue')

plt.show()
```

# Deconvolution Flow-Matching

```python
import torch
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from DynGenModels.models.mlp import MLP
from DynGenModels.configs.toys_configs import Toys_FlowMatch_MLP_Configs as Configs

#...define hyperparameters:

configs = Configs(data_name = 'smeared_gaussians',
                features = ['x', 'y'],
                num_points = 30000,
                sigma = 1e-6,
                noise_cov = [[0.1,0],[0,1]],
                data_split_fracs = [0.6, 0.1, 0.3],
                epochs = 300,
                early_stopping=1000,
                batch_size = 512,
                warmup_epochs = 50,
                print_epochs = 50,
                lr = 1e-3,
                dim_hidden = 16, 
                device = 'cpu',
                solver='midpoint',
                num_sampling_steps=1000,
                seed = 12345
                )
                
configs.set_workdir(path='../results', save_config=True)
```

```python
from DynGenModels.datamodules.toys.datasets import SmearedGaussDataset as DataSet
from DynGenModels.datamodules.toys.dataloader import ToysDataLoader as Loader

dataset = DataSet(configs)
dataloader = Loader(dataset, configs)
```

```python
from DynGenModels.dynamics.deconvolution import DeconvolutionFlowMatching
from DynGenModels.trainer.trainer import DynGenModelTrainer

mlp = MLP(configs)
cfm = DynGenModelTrainer(dynamics=DeconvolutionFlowMatching(mlp, configs), 
                         dataloader=dataloader,
                         configs=configs)

cfm.train()
```

sampling pipeline:

```python
from DynGenModels.pipelines.SamplingPipeline import FlowMatchPipeline as Pipeline
pipeline = Pipeline(trained_model=cfm, configs=configs)
```

```python

def plot_gaussians(data, num_points=None, title=None):

    import scipy.stats as stats

    fig, ax = plt.subplots(1,1,figsize=(2,2))
    x = data[:num_points,0]
    y = data[:num_points,1]
    z = stats.gaussian_kde(np.vstack([x,y]))(np.vstack([x,y]))
    plt.scatter(x, y, c=z, s=0.2, cmap='jet')
    plt.xlim(-5,5)
    plt.ylim(-5,5)
    plt.title(title, fontsize=6)
    plt.xticks(fontsize=6)
    plt.yticks(fontsize=6)
    plt.tight_layout()
    plt.show()


plot_gaussians(dataset.source, title='noisy data (source)', num_points=len(pipeline.target))
plot_gaussians(pipeline.target, title='deconvoluted data (target)')
plot_gaussians(dataset.truth, title='clean data (truth)', num_points=len(pipeline.target))
```

```python
#...plot trajectories:

fig, ax = plt.subplots(1,1, figsize=(4,4))
for i in range(configs.num_sampling_steps): 
    plt.scatter(pipeline.trajectories[i][:,0], pipeline.trajectories[i][:,1], s=0.0225, color='gray', alpha=0.3)
plt.scatter(pipeline.trajectories[0][:,0], pipeline.trajectories[0][:,1], s=0.3, color='red', alpha=0.6)
plt.scatter(pipeline.trajectories[-1][:,0], pipeline.trajectories[-1][:,1], s=0.3, color='blue', alpha=0.6)
plt.show()
```

# Deconvolution Normalizing Flows

```python
import torch
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from DynGenModels.models.maf import MAFPiecewiseRQS
from DynGenModels.configs.toys_configs import Toys_NormFlow_MAF_RQS_Configs as Configs

#...define hyperparameters:

configs = Configs(data_name = 'smeared_gaussians',
                features = ['x', 'y'],
                num_points = 30000,
                noise_cov = [[0.1,0],[0,1]],
                data_split_fracs = [0.8, 0.19, 0.01],
                epochs = 1000,
                early_stopping=30,
                batch_size = 256,
                warmup_epochs = 20,
                print_epochs = 10,
                lr = 5e-4,
                dim_hidden = 128, 
                device = 'cpu',
                seed = 12345,
                num_transforms = 5,
                num_gen_samples = 6000,
                num_mc_draws = 30
                )

configs.set_workdir(path='../results', save_config=True)
```

```python
from DynGenModels.datamodules.toys.datasets import SmearedGaussDataset
from DynGenModels.datamodules.toys.dataloader import ToysDataLoader

dataset = SmearedGaussDataset(configs)
dataloader = ToysDataLoader(dataset, configs)
```

```python
from DynGenModels.dynamics.deconvolution import DeconvolutionNormFlows
from DynGenModels.trainer.trainer import DynGenModelTrainer

maf = MAFPiecewiseRQS(configs)
nflow = DynGenModelTrainer(dynamics=DeconvolutionNormFlows(maf, configs), 
                           dataloader=dataloader,
                           configs=configs)

nflow.train()
```

```python
from DynGenModels.pipelines.SamplingPipeline import NormFlowPipeline
pipeline = NormFlowPipeline(trained_model=nflow, configs=configs, num_gen_samples=6000)
```

```python
def plot_gaussians(data, num_points=None, title=None):

    import scipy.stats as stats

    fig, ax = plt.subplots(1,1, figsize=(3,3))
    x = data[:num_points,0]
    y = data[:num_points,1]
    z = stats.gaussian_kde(np.vstack([x,y]))(np.vstack([x,y]))
    plt.scatter(x, y, c=z, s=2, cmap='jet')
    plt.xlim(-4,4)
    plt.ylim(-4,4)
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
plot_gaussians(dataset.source, num_points=6000, title='noisy data (source)')
plot_gaussians(pipeline.target, title='deconvoluted data (target)')
plot_gaussians(dataset.truth, num_points=6000, title='clean data (truth)')
```
